### Define

```python
# Check the values in the columns 
print(master_df.doggo.value_counts())
print(master_df.floofer.value_counts())
print(master_df.pupper.value_counts())
print(master_df.puppo.value_counts())
```

### Code

```python
# Select the columns to melt and to remain
columns_to_melt = ['doggo', 'floofer', 'pupper', 'puppo']
columns_to_stay = [x for x in master_df.columns.tolist() if x not in columns_to_melt]

# Melt the the columns into values
master_df = pd.melt(master_df, id_vars = columns_to_stay, value_vars = columns_to_melt, 
                         var_name = 'stages', value_name = 'dog_stage')

# Delete column 'stages'
master_df = master_df.drop('stages', 1)

```
